```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
Base_Dados = pd.read_csv('PS4_GamesSales.csv', encoding='latin-1')

Base_Dados.head()
```

```python
Base_Dados.isnull().sum()
```

```python
plt.figure( figsize=(14,5))
plt.title('Verificando campos nulos')
sns.heatmap( Base_Dados.isnull(), cbar = False );
```

```python
Base_Dados.dropna(inplace = True)
```

```python
Base_Dados.describe()
```

```python
plt.figure( figsize=(10,5) )
plt.title('Quantidade de vendas Globais (mi)', loc='left', fontsize=14)
sns.barplot( data=Base_Dados, x='Year', y='Global', ci=None, color='#008080', estimator = sum)
plt.ylabel('Quantidade de vendas (mi)');
```

```python
Base_Dados = Base_Dados.loc[ ( Base_Dados['Year'] != 2019 ) & ( Base_Dados['Year'] != 2020 ) ]

Base_Dados.head()
```

```python
plt.figure(figsize=(14,5))
plt.style.use('ggplot')
plt.title('Distribuio de vendas globais', loc='left', fontsize=14)
sns.kdeplot(Base_Dados['Global'], shade=True, bw=1, color='#008080', linewidth=2.5);
```

```python
Base_Dados.groupby(by = ['Year'] ).sum()
```

```python
plt.figure(figsize = (10,4 ) )
plt.title('Analise da distribuio global (mi)', loc='left', fontsize = 14 )
sns.boxplot(data=Base_Dados, x='Year', y='Global')
```

```python
Base_Dados.loc[Base_Dados ['Global'] >=10]
```

```python
Base_Dados
```

```python
Analise = Base_Dados.groupby( by=['Year'] ).sum().reset_index()

America = [ America / Total * 100 for America, Total in zip( Analise['North America'], Analise['Global'] ) ]
Europe = [ Europe / Total * 100 for Europe, Total in zip( Analise['Europe'], Analise['Global'] ) ]
Japan = [ Japan / Total * 100 for Japan, Total in zip( Analise['Japan'], Analise['Global'] ) ]
Global = [ Global / Total * 100 for Global, Total in zip( Analise['Global'], Analise['Global'] ) ]

America, Europe, Japan, Global
```

```python
Analise
```

```python
# Tamanho
plt.figure( figsize=(10, 5) )

# LArgura barra no grfico
Largura_Barra = 0.85
Rotulos = Analise['Year']
Grupos = [0, 1, 2, 3, 4, 5]

# titulo
plt.title('Anlise distribuio por continentes')

# Plot da America
plt.bar( Grupos, America, width=Largura_Barra, color='#b5ffb9', edgecolor='white' )

# Plot da Europa
plt.bar( Grupos, Europe, bottom=America, width=Largura_Barra, color='#f9bc86', edgecolor='white' )

# Plot do Japao
plt.bar( Grupos, Japan, bottom=[ A + B for A, B in zip(America, Europe) ], width=Largura_Barra, color='#a3acff', edgecolor='white' )

# Plot do Resto do mundo
plt.bar( Grupos, Global, bottom=[ A + B + C for A, B, C in zip(America, Europe, Japan) ], width=Largura_Barra, color='#d3acfe', edgecolor='white' )

# Labels
plt.xticks( Grupos, Rotulos )
plt.xlabel('Grupo')
plt.ylabel('Distribuio %')

# Legenda
plt.legend( ['America N', 'Europa',' Japo', 'Mundo'], loc='upper left', bbox_to_anchor=(0.15, -0.1), ncol=4 );
```

```python
from sklearn.preprocessing import LabelEncoder
Funcao_Label = LabelEncoder()

Base_Dados ['Produtor'] = Funcao_Label.fit_transform(Base_Dados['Publisher'] )
Base_Dados ['Genero'] = Funcao_Label.fit_transform(Base_Dados['Genre'] )
Base_Dados ['Jogo'] = Funcao_Label.fit_transform(Base_Dados['Game'] )

Base_Dados.head()
```

```python
Paleta_Cores = sns.color_palette('husl', 8)
Paleta_Cores
```

```python
plt.figure(figsize=(20,5))
plt.title('Analise dos generos de game (mi)')
sns.scatterplot(data=Base_Dados, x='Genre', y='Global', color=Paleta_Cores[0])
```

```python
plt.figure(figsize=(20,5))
plt.title('Analise dos Produtores de game (mi)')
sns.scatterplot(data=Base_Dados, x='Produtor', y='Global', color=Paleta_Cores[0])
```

```python
plt.figure( figsize=(20,5))
plt.title('Anlise dos Games (mi)')
sns.scatterplot(data=Base_Dados, x='Jogo', y='Global', color=Paleta_Cores[0] );
```
